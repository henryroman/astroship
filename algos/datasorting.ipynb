```python

# Step 1: Load the filenames into a list
filenames = [
    "12-12-clone.jpg",
    "12-12.jpg",
    "3rdcoastgenetics logo.png",
    "Ajo Blanco.jpg",
    "amaretto-mintz-clone.jpg",
    "Amaretto-Mintz.PNG",
    "Apple Mochi.MOV",
    "Bacio Gelato.HEIC",
    "bacio-gelato-teen.jpg",
    "Bacio-Gelato.jpg",
    "Beach Body.jpg",
    "betterteensvideo.mp4",
    "Black Bubblegum.jpg",
    "Black Trufflez.jpg",
    "black-bubblegum-horatio-clone.jpg",
    "Buttafingerz.jpg",
    "Cadillac Rainbowz.jpeg",
    "Calamus Nursery Logo (2).png",
    "CalamusNurseryMenu.pdf",
    "Carbon Candy.jpg",
    "Carbon Fiber Outdoor.mp4",
    "Carbon-Candy-pre-teen.jpg",
    "carbon-candy-teen.jpg",
    "carbon-fibre-pre-teen.jpg",
    "carbon-fibre-teen.jpg",
    "CC (1).png",
    "Cherry Gass.png",
    "clone.jpg",
    "Detroit Breath.jpeg",
    "favicon.svg",
    "gelato-33-teen.jpg",
    "Gelato33-flower.jpg",
    "glitterbomb-clone.jpg",
    "Glitterbomb-pre-teen.jpg",
    "glitterbomb-teen.jpg",
    "Glitterbomb.JPG",
    "goodGreenhouseVideo.mp4",
    "goodQuantityExample.jpg",
    "Gush-Mints.jpg",
    "Happy-Dreams-Genetics.png",
    "hollywood-pre-teen.jpg",
    "hollywood-teen.jpg",
    "hollywood.jpg",
    "homepageimage.png",
    "Ice cream cake.jpg",
    "Ice cream cake.MOV",
    "ice-cream-cake-clone.jpg",
    "ice-cream-cake-teen.jpg",
    "IMG_8693.JPG",
    "Key Lime Jack.jpg",
    "LCG.HEIC",
    "Lemon Cherry Gelato.jpg",
    "lemon-cherry-gelato-clone.jpg",
    "malibu-clone.jpg",
    "OG18.jpg",
    "Oreoz.JPG",
    "Paradise Snacks.jpeg",
    "Permanent Cherries.jpg",
    "Permanent Marker-pre-teen.jpg",
    "Permanent Marker.jpg",
    "Permanent push Pop.jpeg",
    "permanent-gas-clone.jpg",
    "permanent-gas-teen.jpg",
    "Pink-runtz.heic",
    "Pink-runtz.jpg",
    "Pixey Stick.jpeg",
    "Pure Michigan (1).jpg",
    "Pure Michigan.JPG",
    "robots.txt",
    "runtz-huge-teen.jpg",
    "runtz-teen.jpg",
    "Runtz.WEBP",
    "Scented Marker.JPG",
    "scented-marker-huge-teen.jpg",
    "scented-marker-teen.jpg",
    "scented-marker-teenbigger.jpg",
    "super-pop-rocks-clone.jpg",
    "Taffy Twist.jpeg",
    "xenon-clone-better.jpg",
    "xenon-clone.jpg",
    "zoap-clone.jpg",
    "zoap-teen-even-bigger.jpg",
    "zoap-teen.jpg",
    "Zoap.jpeg",
    "zoap.jpg"
]

# Step 2: Define the products list (full JSON data)
products = [
    {
        "slug": "12-12",
        "name": "12:12 (Sativa)",
        "price": "$48",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "London Pound Cake x OG Fruit Smoothie",
        "images": []
    },
    {
        "slug": "ajo-blanco",
        "name": "Ajo Blanco (Indica)",
        "price": "$35",
        "description2": "9-10 Weeks",
        "description3": "High Yielder",
        "description4": "Average THC 30%",
        "description5": "Sweet, Garlic, Citrus, Green",
        "description": "Garlic Pancakes x Vanilla Cheesecake",
        "images": []
    },
    {
        "slug": "amaretto-mintz",
        "name": "Amaretto Mintz (Hybrid)",
        "price": "$89",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 28%",
        "description5": "Sweet, Berry, Minty, Slightly purple",
        "description": "Cherry Punch x Animal Mintz",
        "images": []
    },
    {
        "slug": "bacio-gelato",
        "name": "Bacio Gelato (Hybrid)",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 30%",
        "description5": "Berry, Creamy, Citrusy, Hints of purple",
        "description": "Sunset Sherbet x Thin Mint GSC",
        "images": []
    },
    {
        "slug": "black-bubblegum",
        "name": "Black Bubblegum (Indica)",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "High Yielder",
        "description4": "Average THC 33%",
        "description5": "Candy, Citrusy, Gas, Purple",
        "description": "Spritzer x Horchata",
        "images": []
    },
    {
        "slug": "carbon-candy",
        "name": "Carbon Candy (Hybrid)",
        "price": "$35",
        "description2": "9-10 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 30%",
        "description5": "Candy, Fruity, Gas, Purple",
        "description": "Carbon Fiber x Gushers",
        "images": []
    },
    {
        "slug": "cherry-gass",
        "name": "Cherry Gass (Indica)",
        "price": "$48",
        "description2": "8-9 Weeks",
        "description3": "High Yielder",
        "description4": "Average THC 30%",
        "description5": "Sweet, Cherry, Gassy, Dark Green",
        "description": "(Triple OG x Humboldt Gelato) x Cherry Frosting",
        "images": []
    },
    {
        "slug": "key-lime-jack",
        "name": "Key Lime Jack (Indica)",
        "price": "$35",
        "description2": "9-10 Weeks",
        "description3": "High Yielder",
        "description4": "Average THC 28%",
        "description5": "Sweet, Sour, Skunky, Green",
        "description": "Bling Cheesecake x Jack Herer",
        "images": []
    },
    {
        "slug": "paradise-snacks",
        "name": "Paradise Snacks (Indica)",
        "price": "$48",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 42%",
        "description5": "Candy, Berry, Pine, Slightly purple",
        "description": "Runtz Mints 50 x OG Fruit Smoothie",
        "images": []
    },
    {
        "slug": "permanent-cherries",
        "name": "Permanent Cherries (Indica)",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 30%",
        "description5": "Candy, Fruity, Gas, Slightly purple",
        "description": "Lemon Cherry Gelato x Kush Mints bx1 x Permanent Marker",
        "images": []
    },
    {
        "slug": "permanent-marker",
        "name": "Permanent Marker (Indica)",
        "price": "$89",
        "description2": "8-9 Weeks",
        "description3": "High Yielder",
        "description4": "Average THC 30%",
        "description5": "Candy, Gelato, Gas, Purple",
        "description": "Biscotti x Jealousy x Sherbet bx",
        "images": []
    },
    {
        "slug": "lcg",
        "name": "Lemon Cherry Gelato (Indica)",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 28%",
        "description5": "Candy, Berry, Gassy, Hints of purple",
        "description": "(Zkittlez x Gelato 41) x (Gelato 33 x LPC)",
        "images": []
    },
    {
        "slug": "zoap",
        "name": "Zoap (Hybrid)",
        "price": "$48",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 28%",
        "description5": "Fruity, Creamy, Skunky, Purple",
        "description": "Pink Guava #16 X Rainbow Sherbet",
        "images": []
    },
    {
        "slug": "runtz",
        "name": "Runtz (Hybrid)",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium Yielder",
        "description4": "Average THC 28%",
        "description5": "Candy, Fruity, Gas, Slightly purple",
        "description": "Zkittlez x Gelato",
        "images": []
    },
    {
        "slug": "scented-marker",
        "name": "Scented Marker (Indica)",
        "price": "$89",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 30%",
        "description5": "Candy, Tropical, Gas, Green",
        "description": "Pineapple Fuzz X Permanent Marker",
        "images": []
    },
    {
        "slug": "black-trufflez",
        "name": "Black Trufflez",
        "price": "$48",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Tall slender porcelain bottle with natural clay textured body and cork stopper.",
        "images": []
    },
    {
        "slug": "buttafingerz",
        "name": "Buttafingerz",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Olive drab green insulated bottle with flared screw lid and flat top.",
        "images": []
    },
    {
        "slug": "cadillac-rainbowz",
        "name": "Cadillac Rainbowz",
        "price": "$89",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Person using a pen to cross a task off a productivity paper card.",
        "images": []
    },
    {
        "slug": "detroit-breath",
        "name": "Detroit Breath",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Olive drab green insulated bottle with flared screw lid and flat top.",
        "images": []
    },
    {
        "slug": "glitterbomb",
        "name": "Glitterbomb",
        "price": "$89",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Person using a pen to cross a task off a productivity paper card.",
        "images": []
    },
    {
        "slug": "gush-mints",
        "name": "Gush Mints",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Hand holding black machined steel mechanical pencil with brass tip and top.",
        "images": []
    },
    {
        "slug": "og-18",
        "name": "OG-18",
        "price": "$89",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Person using a pen to cross a task off a productivity paper card.",
        "images": []
    },
    {
        "slug": "oreoz",
        "name": "Oreoz",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Hand holding black machined steel mechanical pencil with brass tip and top.",
        "images": []
    },
    {
        "slug": "permanent-push-pop",
        "name": "Permanent Push Pop",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Hand holding black machined steel mechanical pencil with brass tip and top.",
        "images": []
    },
    {
        "slug": "pink-runtz",
        "name": "Pink Runtz",
        "price": "$48",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Tall slender porcelain bottle with natural clay textured body and cork stopper.",
        "images": []
    },
    {
        "slug": "pixey-stick",
        "name": "Pixey Stick",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Olive drab green insulated bottle with flared screw lid and flat top.",
        "images": []
    },
    {
        "slug": "pure-michigan",
        "name": "Pure Michigan",
        "price": "$89",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Person using a pen to cross a task off a productivity paper card.",
        "images": []
    },
    {
        "slug": "taffy-twist",
        "name": "Taffy Twist",
        "price": "$35",
        "description2": "8-9 Weeks",
        "description3": "Medium/High Yielder",
        "description4": "Average THC 35%",
        "description5": "Gas, Sweet, Cream, Slightly purple",
        "description": "Hand holding black machined steel mechanical pencil with brass tip and top.",
        "images": []
    }
]

# Step 3: Create a function to match images to products
def match_images_to_products(filenames, products):
    import re

    # Preprocess filenames to make matching easier
    filenames_lower = [fname.lower() for fname in filenames]

    # Build a lookup for products by slug and name
    product_lookup = {}
    for product in products:
        # Slug
        slug = product['slug'].lower()
        product_lookup[slug] = product

        # Name without parenthesis and descriptors
        name = re.sub(r'\(.*\)', '', product['name']).strip().lower()
        name = re.sub(r'\W+', '-', name)  # Replace non-alphanumeric with hyphens
        product_lookup[name] = product

    # Match filenames to products
    for fname_lower, fname_original in zip(filenames_lower, filenames):
        matched = False
        for key in product_lookup:
            if key in fname_lower:
                product = product_lookup[key]
                product['images'].append(fname_original)
                matched = True
                break
        if not matched:
            # Handle unmatched images if necessary
            pass

    return products

# Step 4: Update the products' images field
updated_products = match_images_to_products(filenames, products)

# Step 5: Display the updated products list
import json
print(json.dumps(updated_products, indent=4))
```
